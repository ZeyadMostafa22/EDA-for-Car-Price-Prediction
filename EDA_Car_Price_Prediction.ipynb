**correct the company names**

```python
company_name_mapping = {
    'maxda': 'mazda',
    'Nissan': 'nissan',
    'porcshce': 'porsche',
    'toyouta': 'toyota',
    'vokswagen': 'volkswagen',
    'vw': 'volkswagen'
}

data['Company'].replace(company_name_mapping, inplace=True)
```

**Average Car Prices by Company**

```python
company_prices = data.groupby("Company")["price"].mean().reset_index()
```

```python
company_prices = company_prices.sort_values(by="price", ascending=False)
```

```python
plt.figure(figsize=(15, 6))
sns.barplot(data=company_prices, x="price", y="Company", palette="viridis")
plt.title("Average Car Prices by Company")
plt.xlabel("Average Price")
plt.ylabel("Company")
plt.show()
```

```python
correlation_matrix = data.corr(numeric_only=True)

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
